```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3

conn = sqlite3.connect('covid-19.db')
cur = conn.cursor()

strSQL = "SELECT * FROM "
strSQL = strSQL + " US_covid19_vaccination  "

df = pd.read_sql(strSQL, conn)

conn.close()
```

```python
df
```

```python
df = df.groupby('date').sum()
df
```

```python
df.info()
```

```python
%matplotlib inline
```

```python
fig = plt.figure(figsize = (300,100)) 

x = df['cases']
y = df['vaccinated']

plt.plot(x, label = "covid cases", linewidth = 30)
plt.plot(y, label = "people fully vaccinated", linewidth = 30)
plt.legend(loc = "best", frameon = True, fontsize = 200) 
plt.show()
```

```python
daily_covid_case = []
daily_covid_case.append(None)

for i in range(len(df)-1):
    daily_covid_case.append(df.iloc[i+1, 1] - df.iloc[i, 1])

df['daily_covid_case'] = daily_covid_case
df
```

```python
daily_vaccination = []
daily_vaccination.append(None)

for i in range(len(df)-1):
    daily_vaccination.append(df.iloc[i+1, 2] - df.iloc[i, 2])

df['daily_vaccination'] = daily_vaccination
df
```

```python
df.to_csv("daily_cases.csv")
```

```python
fig = plt.figure(figsize = (300,100)) 
x = df['daily_covid_case']

plt.xlabel('date')
plt.ylabel('daily covid case')
plt.ylim([0, 200000])    

plt.plot(x, label = "daily covid cases", linewidth = 30)
plt.legend(loc = "best", frameon = True, fontsize = 300) 
plt.show()
```

```python
fig = plt.figure(figsize = (300,100)) 
x = df['daily_vaccination']

plt.xlabel('date')
plt.ylabel('daily vaccination case')
plt.ylim([0, 2000000])    

plt.plot(x, label = "daily vaccination cases", linewidth = 30)
plt.legend(loc = "best", frameon = True, fontsize = 300) 
plt.show()
```
