```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3

conn = sqlite3.connect('covid-19.db')
cur = conn.cursor()

strSQL = "SELECT * FROM "
strSQL = strSQL + " US_covid19_vaccination  "

df = pd.read_sql(strSQL, conn)

conn.close()
```

```python
df
```

```python
seen = set()
state = []
for i in df['state']:
  if i not in seen:
    seen.add(i)
    state.append(i)
    
print(state)
```

```python
i = input("state:   ")

df.loc[df['state'] == i]
```

```python
fig = plt.figure() 

for i in state:
    x = df.loc[df['state'] == i]['cases']

    plt.plot(x, label = "covid cases", linewidth = 3)
    plt.title(i)
    plt.legend(loc = "best", frameon = True) 
    plt.show()
```

```python
fig = plt.figure() 

for i in state:
    x = df.loc[df['state'] == i]['people_fully_vaccinated']

    plt.plot(x, label = "people fully vaccinated", linewidth = 3)
    plt.title(i)
    plt.legend(loc = "best", frameon = True)
    plt.show()
```
