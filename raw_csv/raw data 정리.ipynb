```python
import pandas as pd

covid_cases = pd.read_csv('us-states.csv')
vaccination = pd.read_csv('us_state_vaccinations.csv')
```

```python
covid_cases
```

```python
covid_cases = covid_cases[['date','state','cases']]
```

```python
covid_cases
```

```python
vaccination = vaccination[['date','location','people_fully_vaccinated','people_fully_vaccinated_per_hundred']]
```

```python
vaccination
```

```python
merged = pd.merge(covid_cases, vaccination, on = 'date', how = 'outer')
```

```python
merged
```

```python
final = merged[merged['state'] == merged['location']]
```

```python
final = final.drop(['location'], axis = 1)
```

```python
final
```
